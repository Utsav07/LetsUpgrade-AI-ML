```python
import pandas as pd
dataset = pd.read_csv('general_data.csv')
```

```python
dataset.head()
```

```python
dataset["Attrition"] = pd.get_dummies(dataset["Attrition"])
dataset["BusinessTravel"] = pd.get_dummies(dataset["BusinessTravel"])
dataset["Department"] = pd.get_dummies(dataset["Department"])
dataset["EducationField"] = pd.get_dummies(dataset["EducationField"])
dataset["Gender"] = pd.get_dummies(dataset["Gender"])
dataset
```

```python
def Hypothesis_decision(p):
    if p > 0.05:
        print("Null hypothesis is accepted!")
    else:
        print("Null hypothesis is rejected!")
```

```python
from scipy.stats import mannwhitneyu
a2 = dataset.DistanceFromHome
a1 = dataset.Attrition
stat,p = mannwhitneyu(a1, a2)
print(stat, p)
Hypothesis_decision(p)
```

```python
from scipy.stats import mannwhitneyu
a1 = dataset.TotalWorkingYears
a2 = dataset.Attrition
stat,p = mannwhitneyu(a1, a2)
print(stat, p)
Hypothesis_decision(p)
```
